```python
#  Dataを取得
root_path = "/content/drive/MyDrive/jojo_poser"

from PIL import Image
!pip install pyheif
import pyheif
import glob

def conv(image_path):
    new_name = image_path.replace('HEIC', 'png')
    print(new_name)
    assert image_path!=new_name
    heif_file = pyheif.read(image_path)
    data = Image.frombytes(
        heif_file.mode,
        heif_file.size,
        heif_file.data,
        "raw",
        heif_file.mode,
        heif_file.stride,
        )
    data.save(new_name, "PNG")

for phase in ["train"]:
  root_dir = root_path+"/images/"+phase
  for label in LABELS:
    path = root_dir + "/" + label + "/IMG_6205.HEIC"
    print(path)
    lst = glob.glob(path)
    for l in lst:
        conv(l)
```
