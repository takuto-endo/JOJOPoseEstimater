```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("log_output2.csv")
df.head()
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('epoch_num')  # x軸ラベル
ax.set_ylabel('loss')  # y軸ラベル
ax.plot(df['epoch'], df['train_loss'], label='train_loss', linestyle='solid')
ax.plot(df['epoch'], df['val_loss'], label='val_loss', linestyle='solid')
ax.legend(loc=0)
fig.tight_layout()
plt.show()
```
